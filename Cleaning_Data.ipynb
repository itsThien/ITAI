```python
# Importing labraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Excel file into pandas DataFrame
df = pd.read_excel(r"/content/Customer Call List.xlsx")
df
```

```python
# Remove duplicate rows from the DF
df = df.drop_duplicates()
df
```

```python
# Drop the Not_Useful_Column
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# Clean the Last_Name column by removing unwanted characters
########################################################################################

df["Last_Name"] = df["Last_Name"].str.strip("/")
df["Last_Name"] = df["Last_Name"].str.strip("/_")
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
# Cleans and standardizes the Phone_Number column.
# Removes all non-alphanumeric characters using a regular expression ([^a-zA-Z0-9]).
# Formats the remaining numbers into a xxx-xxx-xxxx format.
# Finally, it removes the empty strings that resulted from converting NaN and N/a
# values to strings, which appeared as nan-- and Na--.
########################################################################################

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
# Splits the Address column into three new columns: Street_Address, State, and Zip_Code.
# Fillna('') is used to replace any NaN values resulting from the split with
# empty strings.
# The output shows the DataFrame with the three new, separated address columns.
########################################################################################

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True).fillna('')
df
```

```python
# Standardizes the Do_Not_Contact column.
# Replaces Yes with Y and No with N to
# create a consistent representation for contact status.
########################################################################################

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
# Replaces inconsistent values across the entire DataFrame.

df = df.replace('N/a','')
df = df.replace('NaN','')


df=df.fillna('')
df
```

```python
# Filters the DataFrame to remove rows where the Do_Not_Contact column is 'Y'.

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
# Further refines the DataFrame by removing rows with empty Phone_Number values.

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

```python
# Resets the index of the final DataFrame.
df = df.reset_index(drop=True)
df
```
